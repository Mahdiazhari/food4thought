```python
import pandas as pd

from  util.spider import Spider
```

```python
attrs = {
        'name'       :  'name',
        'ingredients':  'recipeIngredient',
        'total_time':   'totalTime',
        'instructions': 'recipeInstructions',
        'servings':     'recipeYield',
        'category':     'recipeCategory',
        'prep_time':    'prepTime',
        'cook_time':    'cookTime',
}
available_json = {'xpath' : "normalize-space(//script[@type='application/ld+json'][contains(text(), 'recipeIngredient')])"}

listing={'items': '//div[@class="resept_thumb"]/a/@href', 'next': { 'next_page_str': '?pg={}', 'type': 'url'}}
seeds = ['https://tagamlar.com/resept/']
```

```python
custom_header = { #setup custom header for russian script
        'referer': 'https://www.google.com/',
        'Accept-Language': '*',
        'Accept-Encoding': '*',
        'Accept': '*',
        'user-agent':''}
```

```python
tagamlar_spider= Spider('https://tagamlar.com', seeds= seeds, listing =listing,attrs= attrs, available_json=available_json,header=custom_header)
```

```python
tagamlar_spider.scrape_one_item('https://tagamlar.com/resept/towukly-maklube-multiwarkada/')
```

```python
result_list = tagamlar_spider.start_scrape(multithread=False)
```

```python
df = pd.DataFrame(result_list)
```
