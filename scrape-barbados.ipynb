```python
df_copy = df.copy()
```

```python
df_copy.replace("", None, inplace=True)
```

```python
df_copy
```

```python
df_copy.dropna(subset=['name'], inplace=True)
```

```python
df_copy
```

```python
#filtering snippet
exclude_list = ['Desserts', 'Beverages', 'Cocktails']
def exclude_cat(exclusions, list):
    if list:
        for item in list:
            if item in exclusions: return True
        return False
    else: return False
```

```python
exclude_cat(exclude_list, df_copy.iloc[0]['category'][0])
```

```python
df_copy_filtered = df_copy[~(df_copy.apply(lambda x: exclude_cat(exclude_list, x['category'],),axis=1))]
```
