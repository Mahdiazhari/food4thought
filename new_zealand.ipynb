```python
import pandas as pd
import json
import requests
import random
from lxml import html
import re

from  util.spider import Spider
# from util.http_utility import get_http_headers
# from util.user_agent import get_random_ua
```

```python
attrs = {
        'name':         '//meta[@name="og:title"]/@content',
        'ingredients':  'normalize-space(//ul[@class="recipe-ingredients__list"])',
        'total_time':   'normalize-space(//div[@class="recipe-facts__details recipe-facts__time"])',
        'instructions': 'normalize-space(//ul[@class="recipe-directions__steps"])',
        'servings':     'normalize-space(//div[@class="recipe-facts__details recipe-facts__servings"])',
        'category':     'normalize-space(//span[@class="recipe-breadcrumbs__text category"])',
        'prep_time':    '',
        'cook_time':    '',
}

listing={'items': '//div[contains(@class,"fd-tile fd-recipe")]//h2//a/@href', 'next': { 'next_page_str': '?pn={}', 'type': 'url'}}
seeds = ['https://www.food.com/topic/new-zealand']
```

```python
nz_spider =  Spider('https://www.food.com/', seeds= seeds, listing =listing,attrs= attrs)
```

```python
nz_spider.scrape_one_item('https://www.food.com/recipe/fluffernutter-181487')
```

```python
result_list = nz_spider.start_scrape(max_pages=130) #max page is only 130
```

```python
result_df = pd.DataFrame(result_list)
```
